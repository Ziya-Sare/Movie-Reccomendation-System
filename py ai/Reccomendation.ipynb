```python
import numpy as np
import pandas as pd
```

```python
column_names = ['admin_id', 'movie_id', 'rating']
df = pd.read_csv('admin.data', sep='\t', names=column_names)
```

```python
movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles.head()
```

```python
len(movie_titles)
```

```python
df = pd.merge(df, movie_titles, on='movie_id')
df.head()
```

```python
moviemat = df.pivot_table(index='admin_id',columns='name',values='rating')
moviemat.head()
```

```python
type(moviemat)
```

```python
ratings = moviemat["Dead Man Walking (1995)"]
ratings.head()
```

```python
similar= moviemat.corrwith(ratings)
```

```python
len(df)
```

```python
corr_ = pd.DataFrame(similar, columns=['Correlation'])
corr_.dropna(inplace=True)
corr_.head()
```

```python
corr_.sort_values('Correlation',ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('name')['rating'].mean())
ratings.sort_values('rating',ascending=False).head()
```

```python
ratings['rating_oy_sayisi'] = pd.DataFrame(df.groupby('name')['rating'].count())
ratings.head()
```

```python
corr_ = corr_.join(ratings['rating_oy_sayisi'])
corr_.head()
```

```python
corr_[corr_['rating_oy_sayisi']>100].sort_values('Correlation',ascending=False).head()
```

```python
from tkinter import * 

window = Tk()
window.title("MOVİE RECCOMENDS")
window.geometry("1500x750")
window.configure(background='black')
window.wm_attributes('-alpha',0.85)

lb_baslik=Label(text="Movie Reccomendation System",font="helvetica 30",fg="#13ABF2",width=50,bg='black')
lb_baslik.place(x=150,y=30)
lb_baslik=Label(text="Choose a Movie...",font="SansSerif 12",fg="#13ABF2",bg="black")
lb_baslik.place(x=450,y=100)


data=movie_titles['name']
list=Listbox(window,height="10",width="50",bg="black",bd="8px",highlightbackground="#13ABF2",highlightcolor="#13ABF2",highlightthickness=4,font="SansSerif",fg="#13ABF2")
list.place(x=450,y=160)
label=Label()
label.place(x=1100,y=170)

a=1
for i in data:
    list.insert(a,i)
    a=a+1


lb1=Label(text="")
lb1.place(x=450,y=450)
lb=Label(text="")
lb.place(x=450,y=500)


def metot():
    eleman=list.curselection()
    if(eleman):
        deger=list.get(eleman)
        lb1.configure(text=deger,font="SansSerif 12",fg="#13ABF2",bg="black")
        label.configure(text="")
        ratings = moviemat[deger]
        similar = moviemat.corrwith(ratings)
        corr_ = pd.DataFrame(similar, columns=['Correlation'])
        corr_.dropna(inplace=True)
        corr_.sort_values('Correlation',ascending=False)
        ratings = pd.DataFrame(df.groupby('name')['rating'].mean())
        ratings.sort_values('rating',ascending=False)
        ratings['Rating'] = pd.DataFrame(df.groupby('name')['rating'].count())
        ratings.sort_values('Rating',ascending=False)
        corr_.sort_values('Correlation',ascending=False)
        corr_ = corr_.join(ratings['Rating'])
        text1=corr_[corr_['Rating']>100].sort_values('Correlation',ascending=False).head(10)
        
        lb.configure(text=text1, font="SansSerif 12", fg= "#13ABF2",bg = "black",bd="7px",highlightbackground="#13ABF2",justify = RIGHT)
        
               
    else:
        label.configure(text="Please Choose!!!", font="SansSerif 15",fg="#13ABF2",bg="black")
        
                
b1=Button(text="Show Reccomendations",command=metot, font= "SansSerif 15", borderwidth=35 , bg="#007FFF",fg="black")
b1.place(x=1100, y=225)

window.mainloop()
```
